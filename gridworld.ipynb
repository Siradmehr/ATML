# Gridworld

### Description

In this exercise we implement an example of the agent-environment interface used in reinforcement learning, called "gridworld".

The world consists of an $n \times m$ grid of squares, indexed by $(i,j)$ with $i = 0, 1, \dots, n-1$ and $j = 0, 1, \dots, m-1$.

The state of the environment consists of the player being in one of the squares.

The possible actions are steps in the directions "UP", "RIGHT", "DOWN", "LEFT".

The rewards and states after each action can depend on multiple factors:
- Some squares can give positive/negative rewards
- Squares might be "blocked" and not possible to step on
- Stepping off the board is impossible
- Invalid moves might give a negative reward as "punishment"
- There might be "portals" that take the player to a distant square, regardless of their action
- There might be deterministic or random effects that change the outcome of actions, e.g. "wind" or "ice".

Variants of this gridworld can be used to illustrate a wide range of concepts and algorithms in reinforcement learning.
For instance, in [Sutton & Barto](http://incompleteideas.net/book/the-book-2nd.html) see:
- Example 3.5, 3.8

- Example 6.5, 6.6

- Example 8.1, 8.3

### Implementation

```python
class GridWorld:
    def __init__(self, height=5, width=5, goal=None):
        self.height = height
        self.width = width
        
        # Goal position (default = bottom right)
        self.goal = goal if goal is not None else (height-1, width-1)
        
        self.reset()

    def reset(self):
        """Reset environment to start state"""
        self.pos = (0, 0)
        self.previous_pos = (0, 0)
        self.total_reward = 0.0
        self.done = False
        return self.pos

    def step(self, action):
        
        if self.done:
            return self.pos, 0.0, True

        self.previous_pos = self.pos

        # Movement
        if action == 'up':
            self.pos = (max(0, self.pos[0] - 1), self.pos[1])
        elif action == 'down':
            self.pos = (min(self.height - 1, self.pos[0] + 1), self.pos[1])
        elif action == 'left':
            self.pos = (self.pos[0], max(0, self.pos[1] - 1))
        elif action == 'right':
            self.pos = (self.pos[0], min(self.width - 1, self.pos[1] + 1))
        else:
            raise ValueError("Invalid action. Use 'up', 'down', 'left', 'right'.")

        # Get reward from the separate function
        reward = self.reward()

        self.total_reward += reward
        self.done = (self.pos == self.goal)

        return self.pos, reward, self.done

    def reward(self):
        """
        This is the method you can easily override or replace.
        Returns reward based on current state transition.
        """
        # Reached goal → big positive reward
        if self.pos == self.goal:
            return 10.0
        
        # Illegal move (bumped into wall)
        if self.pos == self.previous_pos:
            return -1.0
        
        # Normal step
        return 0.0

    def drawworld(self):
        for i in range(self.height):
            for j in range(self.width):
                if (i, j) == self.pos:
                    print('P', end=' ')
                elif (i, j) == self.goal:
                    print('G', end=' ')
                else:
                    print('.', end=' ')
            print()
        print(f"  pos = {self.pos} | total_reward = {self.total_reward:.1f} | done = {self.done}")

    def play(self):
        self.reset()
        print("Interactive mode — goal is G\n")
        
        while not self.done:
            self.drawworld()
            action = input("Action (up/down/left/right/q=quit): ").strip().lower()
            if action in ('q', 'quit'):
                print("Exited.")
                return
            try:
                pos, reward, done = self.step(action)
                if reward != 0:
                    print(f"Reward: {reward:+.1f}")
            except ValueError as e:
                print(e)
        
        self.drawworld()
        print(f"\nGoal reached! Total reward: {self.total_reward:.1f}\n")
```

```python
class GridWorld:
    def __init__(self, 
                 height=6, 
                 width=9, 
                 goal=None, 
                 blocked=None):
        self.height = height
        self.width = width
        
        # Goal
        self.goal = goal if goal is not None else (height-1, width-1)
        
        # Blocked cells: list of (r, c) tuples
        self.blocked = set()
        if blocked:
            for r, c in blocked:
                self.add_blocked(r, c)
        
        # Make sure goal is not blocked
        if self.goal in self.blocked:
            raise ValueError(f"Goal {self.goal} cannot be a blocked cell")
        
        self.reset()

    def reset(self):
        self.pos = (0, 0)
        # Make sure start is not blocked
        if self.pos in self.blocked:
            raise ValueError("Starting position (0,0) is blocked")
        self.previous_pos = self.pos
        self.total_reward = 0.0
        self.done = False
        return self.pos

    def add_blocked(self, row, col):
        """Add a blocked cell. Returns True if successful."""
        if not (0 <= row < self.height and 0 <= col < self.width):
            return False
        if (row, col) == self.goal:
            print("Cannot block the goal position")
            return False
        if (row, col) == (0, 0):
            print("Cannot block the starting position")
            return False
        self.blocked.add((row, col))
        return True

    def remove_blocked(self, row, col):
        """Remove a blocked cell if it exists"""
        self.blocked.discard((row, col))

    def step(self, action):
        if self.done:
            return self.pos, 0.0, True

        self.previous_pos = self.pos

        # Proposed new position
        r, c = self.pos
        if action == 'up':
            r = max(0, r - 1)
        elif action == 'down':
            r = min(self.height - 1, r + 1)
        elif action == 'left':
            c = max(0, c - 1)
        elif action == 'right':
            c = min(self.width - 1, c + 1)
        else:
            raise ValueError("Invalid action")

        new_pos = (r, c)

        # Check if move is allowed
        if new_pos in self.blocked:
            # Stay in place → illegal move
            self.pos = self.previous_pos
        else:
            self.pos = new_pos

        reward = self.reward()
        self.total_reward += reward
        self.done = (self.pos == self.goal)

        return self.pos, reward, self.done

    def reward(self):
        """Override this method or replace it to customize rewards"""
        if self.pos == self.goal:
            return 10.0
        
        # Hit wall or blocked cell
        if self.pos == self.previous_pos:
            return -1.0
        
        return 0.0  

    def drawworld(self):
        for i in range(self.height):
            row = []
            for j in range(self.width):
                cell = (i, j)
                if cell == self.pos:
                    row.append('P')
                elif cell == self.goal:
                    row.append('G')
                elif cell in self.blocked:
                    row.append('#')  
                else:
                    row.append('.')
            print(' '.join(row))
        print(f"  pos = {self.pos} | total = {self.total_reward:.1f} | done = {self.done}")

    def play(self):
        self.reset()
        print("Goal = G    Blocked = #    You = P\n")
        print("Commands: up down left right   q = quit\n")
        
        while not self.done:
            self.drawworld()
            cmd = input("> ").strip().lower()
            if cmd in ('q', 'quit'):
                print("Exited.")
                break
            if cmd in {'up', 'down', 'left', 'right'}:
                try:
                    pos, r, d = self.step(cmd)
                    if r != 0:
                        print(f"reward: {r:+.1f}")
                except ValueError as e:
                    print(e)
            else:
                print("unknown command")
        
        if self.done:
            self.drawworld()
            print(f"\nGoal reached! Total reward: {self.total_reward:.1f}\n")
```

```python
class GridWorld:
    def __init__(self, 
                 height=7, 
                 width=10, 
                 goal=None, 
                 blocked=None, 
                 teleports=None):
        """
        GridWorld with blocked cells and teleports.
        
        Args:
            height, width: grid size
            goal: tuple (row, col) or None → defaults to bottom-right
            blocked: list of (r,c) tuples for impassable cells
            teleports: dict {(r_from,c_from): (r_to,c_to), ...}
        """
        self.height = height
        self.width = width
        
        # Goal position
        self.goal = goal if goal is not None else (height - 1, width - 1)
        
        # Initialize containers FIRST to avoid attribute errors
        self.blocked = set()
        self.teleports = {}     # source → destination
        
        # Add blocked cells
        if blocked:
            for r, c in blocked:
                self.add_blocked(r, c)
        
        # Add teleports
        if teleports:
            for (r1, c1), (r2, c2) in teleports.items():
                self.add_teleport(r1, c1, r2, c2)
        
        # Safety checks
        if self.goal in self.blocked:
            raise ValueError("Goal cannot be a blocked cell")
        if self.goal in self.teleports:
            raise ValueError("Goal cannot be a teleport entrance")
        if (0, 0) in self.blocked:
            raise ValueError("Starting position (0,0) cannot be blocked")
        
        self.reset()

    def reset(self):
        """Reset to starting state"""
        self.pos = (0, 0)
        self.previous_pos = self.pos
        self.total_reward = 0.0
        self.done = False
        return self.pos

    def add_blocked(self, row, col):
        pos = (row, col)
        if not (0 <= row < self.height and 0 <= col < self.width):
            return False
        if pos == self.goal:
            print("Cannot block the goal position")
            return False
        if pos == (0, 0):
            print("Cannot block the starting position")
            return False
        
        if pos in self.teleports or pos in self.teleports.values():
            print(f"Cannot block teleport position: {pos}")
            return False
        self.blocked.add(pos)
        return True

    def remove_blocked(self, row, col):
        self.blocked.discard((row, col))

    def add_teleport(self, r_from, c_from, r_to, c_to):
        src = (r_from, c_from)
        dst = (r_to, c_to)
        
        if not (0 <= r_from < self.height and 0 <= c_from < self.width and
                0 <= r_to   < self.height and 0 <= c_to   < self.width):
            print("Teleport coordinates out of bounds")
            return False
        
        if src == self.goal or dst == self.goal:
            print("Cannot place teleport on goal")
            return False
        if src == (0, 0) or dst == (0, 0):
            print("Cannot place teleport on start")
            return False
        if src in self.blocked or dst in self.blocked:
            print("Cannot place teleport on blocked cell")
            return False
        
        self.teleports[src] = dst
        return True

    def remove_teleport(self, row, col):
        self.teleports.pop((row, col), None)

    def step(self, action):
        if self.done:
            return self.pos, 0.0, True

        self.previous_pos = self.pos

        # Compute intended position
        r, c = self.pos
        if action == 'up':
            r = max(0, r - 1)
        elif action == 'down':
            r = min(self.height - 1, r + 1)
        elif action == 'left':
            c = max(0, c - 1)
        elif action == 'right':
            c = min(self.width - 1, c + 1)
        else:
            raise ValueError("Invalid action. Use: up, down, left, right")

        intended_pos = (r, c)

        
        if intended_pos in self.blocked:
            new_pos = self.pos
        else:
            new_pos = intended_pos

        
        visited = set()
        while new_pos in self.teleports and new_pos not in visited:
            visited.add(new_pos)
            new_pos = self.teleports[new_pos]
            if len(visited) > 50:  # safety
                print("Warning: possible teleport cycle detected")
                break

        self.pos = new_pos

        # Reward calculation
        reward = self._get_reward()
        self.total_reward += reward

        # Check if goal reached
        self.done = (self.pos == self.goal)

        return self.pos, reward, self.done

    def _get_reward(self):
        """You can override this method or change the logic here"""
        if self.pos == self.goal:
            return 10.0
        if self.pos == self.previous_pos:
            return -1.0   # hit wall / blocked / illegal move
        return 0.0     
    def drawworld(self):
        print("-" * (self.width * 2 + 3))
        for i in range(self.height):
            print("|", end=" ")
            for j in range(self.width):
                pos = (i, j)
                if pos == self.pos:
                    print("P", end=" ")
                elif pos == self.goal:
                    print("G", end=" ")
                elif pos in self.blocked:
                    print("#", end=" ")
                elif pos in self.teleports:
                    print("T", end=" ")
                elif pos in self.teleports.values():
                    print("t", end=" ")  # destination
                else:
                    print(".", end=" ")
            print("|")
        print("-" * (self.width * 2 + 3))
        print(f"Position: {self.pos} | Total reward: {self.total_reward:.2f} | Done: {self.done}")

    def play(self):
        """Interactive play mode"""
        self.reset()
        print("Controls: up, down, left, right, quit/q")
        print("Legend: P = you, G = goal, # = blocked, T/t = teleport\n")
        
        while not self.done:
            self.drawworld()
            action = input("Your move: ").strip().lower()
            if action in ["quit", "q"]:
                print("Game ended.")
                break
            if action in ["up", "down", "left", "right"]:
                try:
                    pos, reward, done = self.step(action)
                    if reward != 0:
                        print(f"Reward: {reward:+.2f}")
                except ValueError as e:
                    print(e)
            else:
                print("Invalid input.")
        
        if self.done:
            self.drawworld()
            print("\n*** GOAL REACHED! ***")
            print(f"Final score: {self.total_reward:.2f}\n")
```
